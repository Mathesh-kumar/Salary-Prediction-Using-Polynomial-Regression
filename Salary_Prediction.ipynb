## **Machine Learning model to predict salary for new employee based on position of work using polynomial regression.**
  1. Importing needed libraries
  2. Data preprocessing
  3. Training dataset using simple linear regression
  4. Training dataset using polynomial regression
  5. Visualizing simple linear regression
  6. Visualizing polynomial regression

```
# 1. importing needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
# 2. data preprocessing
# 2.1. importing needed dataset
df = pd.read_csv('Position_Salaries.csv')
df.head()
```

```
# 2.2. splitting dependent & independent variable
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

```
# since our dataset is very small, we are not using train_test_split.
# no feature scaling is needed because of polynomial term.
```

```
# 3. training linear regression on train set.

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
```

```
# 4. training polynomial regression on train set.

from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=2)
X_poly = pf.fit_transform(X)
pf.fit(X_poly, y)
lr_poly = LinearRegression()
lr_poly.fit(X_poly, y)
```

```
# visualizing simple linear regression result

plt.scatter(X, y, color='red')
plt.plot(X, lr.predict(X), color = 'blue')
plt.title('Simple Linear Regression results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```
# visualizing polynomial regression result

plt.scatter(X, y, color='red')
plt.plot(X, lr_poly.predict(pf.fit_transform(X)), color = 'blue')
plt.title('Polynomial Regression results')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
